```python
import nltk
import csv
import numpy as np
import codecs
```

```python
def load_training_data():
    with open("Spam.csv", 'r', encoding="UTF-8", errors="ignore") as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        total_list = []
        X = []
        Y = []

        for row in reader:
            # print(row)
            if row[1] == "neutral":
                continue
            else:
                # print(row[0])
                X.append(row[0])  # Message
                Y.append(row[1])  # Class
                total_list.append(row)

        classes = np.unique(Y)

        for each_class in classes:
            print(each_class)

        print((('-' * 33) + "\n") + "Length of Total List: ", len(total_list))

    return (total_list)


spam_ham = load_training_data()
```

```python
# Text Preprocessing(NLP)
messages=[]
for (words,category) in spam_ham:
    words_filtered=[e.lower() for e in words.split() if len(e)>=3]
    messages.append((words_filtered,category))

# Merge all words from messages into a single list
def get_words_in_messages(messages):
    all_words=[]

    for(words,category) in messages:
        all_words.extend(words)
    return all_words
    
# Collection of unique words
def get_word_features(wordlist):
    wordlist=nltk.FreqDist(wordlist) # Frequency
    word_features=wordlist.keys() # unique words

    return word_features

# Text Document into a feature Dictionary
def extract_features(document):
    document_words=set(document) # set for faster lookup ,remove duplicates and makes checking easier
    features={}

    for word in word_features:
        features["Contains:(%s)" % word]=(word in document_words) # create binary features(True/False)
    return features

word_features=get_word_features(get_words_in_messages(messages))
training_set=nltk.classify.apply_features(extract_features,messages)
classifier=nltk.NaiveBayesClassifier.train(training_set)
    
```
